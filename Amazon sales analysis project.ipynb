```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
import datetime as dt
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby(['Month'])['Total Revenue'].sum().reset_index()
```

```python
yearly_sales = df.groupby(['Year'])['Total Revenue'].sum().reset_index()
```

```python
yearly_month_sales = df.groupby(['Year',"Month"])['Total Revenue'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(data = monthly_sales, x = 'Month',y= 'Total Revenue')
plt.title('Monthly Sales Trends')
plt.show()
```

```python
sns.barplot(data=yearly_sales, x='Year', y='Total Revenue')
plt.title('Yearly Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = yearly_month_sales, x = 'Month',y = 'Total Revenue', hue = 'Year')
plt.title('Yearly_month_wise_trends')
plt.show()
```

```python
#key metrics
```

```python
#financial metrics
```

```python
df['Profit'] = df['Total Revenue'] - df['Total Cost'] 
```

```python
df['profit margin(%)'] = (df['Profit']/df['Total Revenue'])*100
```

```python
#Time based metrics
```

```python
monthly_trends = df.groupby(['Year', 'Month']).agg({
    'Total Revenue': 'sum',
    'Profit': 'sum'
}).reset_index()
print(monthly_trends.head())
```

```python
yearly_trends = df.groupby('Year').agg({
    'Total Revenue': 'sum',
    'Total Cost': 'sum',
    'Profit': 'sum'
}).reset_index()
print(yearly_trends)
```

```python
#growth_rate
df['Revenue Growth(%)'] = df['Total Revenue'].pct_change()*100
```

```python
#shipping metrics
```

```python
df['Shipping Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
average_shipping_duration = df['Shipping Duration'].mean()
print(f"Average Shipping Duration: {average_shipping_duration:.2f} days")
```

```python
sns.lineplot(data=monthly_trends, x='Month', y='Profit', hue='Year', marker='o')
plt.title('Monthly Profit Trends')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

```python
sns.histplot(df['Shipping Duration'], bins=10, kde=True)
plt.title('Shipping Duration Distribution')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.show()
```

```python
#region metrics
```

```python
region_metrics = df.groupby('Region').agg({
    'Total Revenue': 'sum',
    'Profit': 'sum'
}).reset_index()

print(region_metrics)
```

```python
sns.barplot(data=region_metrics, x='Region', y='Total Revenue')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_trends.to_csv('monthly_trends.csv', index=False)
region_metrics.to_csv('region_metrics.csv', index=False)
```
